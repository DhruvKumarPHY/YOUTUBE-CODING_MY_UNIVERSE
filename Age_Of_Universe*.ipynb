```python
# considering the universe has expanded at a constant rate.
import astropy
from astropy import units as u
from astropy import cosmology as cosmo
```

# CASE 1 (WMAP)




__Youtube video for better understanding__ - https://www.youtube.com/watch?v=VCYvLmHx5Nk

```python
from astropy.cosmology import WMAP5
H = WMAP5.H(0)
H
```

```python
H1 = H.to('s-1')
H1
```

```python
Time1 = 1/(H1)
```

```python
Time1 # Age of Universe from WMAP5
```

```python
Time1.to('yr') 
```

# CASE 2 (WMAP7)

```python
from astropy.cosmology import WMAP7
H = WMAP7.H(0)
H
```

```python
H2 = H.to('s-1')
H2
```

 Documentation -  https://docs.astropy.org/en/stable/cosmology/index.html?highlight=Planck

```python
Time2 = 1/H2
```

```python
Time2.to('yr') # Age of Universe by WMAP7 # Most Accurate out of all.
```

# CASE 3 (WMAP9)

```python
from astropy.cosmology import WMAP9
H = WMAP9.H(0)
H
```

```python
H3 = H.to('s-1')
H3
```

```python
Time3 = 1/H3
```

```python
Time3.to('yr')
```

## Let's Take average of all the three values and drive a Average age of Universe.

```python
Avg = (Time1 + Time2 + Time3)/3
print("Average Age of Universe is : " , Avg.to('yr'))
```

# Age of Universe is :  13974414401.94479 yr
